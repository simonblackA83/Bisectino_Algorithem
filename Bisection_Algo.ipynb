```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy
import matplotlib.pyplot as plt
import math
```

```python
####Function f(c) of which one wants to find zero point
def funx(c):
    
    f = (np.sqrt(2*c) * ((2 + c) / 2)) - 12
    return f
```

```python
# Finding a plausible starting interval [a,b]
test_range = np.arange(1,10)   
fx = [funx(i) for i in test_range]

plt.plot(test_range, fx)
zero = np.zeros(len(fx))
plt.plot(test_range, zero, linestyle = '-')
```

```python
#Bisection Algo
a = min(test_range)
b = 1 + max(test_range)
c = (a + b)/2
fx_start = funx(c)

level = 0.000001#Precission Decimals
#1st step
arr = np.array([a,b,c,fx_start]); vals = len(arr)

#Iterations until desired level reached
while abs(fx_start) > level:
    
    if fx_start > 0:
        
        b = c
    
    else:
        
        a = c
    
    c = (a + b)/2
    fx_start = funx(c)
    #n += 1
    add = np.array([a,b,c,fx_start])
    arr = np.concatenate((arr,add))

arr = np.reshape(arr, (int(len(arr)/vals), vals))       
results = pd.DataFrame(arr)
names = ['a','b','c','f(c_n)']
results.columns = names
results
```
